***Analise das Variáveis CO2 e Ligth que no grafico de correlação tiveram alto grau de correlação com a varivael target.***

Ao que parece pelo relatorio elas nao seguem uma distribuição normal. No entanto, não podemos confirmar isso visualmente. Portanto, realizamos o teste de normalidade de Shapiro-Wilk. A hipótese nula (H0) é que os dados são normais.

```python
#teste de normalidade variavel CO2 e Ligth
stat_1, p_1 = stats.shapiro(n_df['CO2'])
stat_2, p_2 = stats.shapiro(n_df['Light'])
print("Estatistica CO2 = %.3f \nValor-p = %.3f" % (stat_1,p_1))
print("Estatistica Light = %.3f \nValor-p = %.3f" % (stat_2,p_2))
```

***Interpretação***

O valor-p é inferior a 0,05, o que implica que a 5% de nível de significância rejeitamos nossa hipótese nula (H0). Portanto, as variáveis não seguem uma distribuição normal.Vou desenhar um grafico de violino para dar uma olhada no comportamento.

```python
ax = sns.violinplot(x='Occupancy',y='CO2',data=n_df,split=True)

#titulo
plt.title("Grafico violino entre as variaveis CO2 e Ocupação", fontsize=18)
```

# Teste de Normalidade

O objetivo desse teste é ajudar a definir se devemos usar testes estatísticos paramétricos (se a variável segue distribuição normal) ou não paramétricos (se variável não segue distribuição normal).


- Hipótese nula: os dados são normalmente distribuídos
- Hipótese alternativa: os dados não são normalmente distribuídos

Se o valor-p for menor que 0,05 (IC de 95%), rejeitamos a hipótese de distribuição normal.

```python
for i in range(len(n_df.columns)-1):
    # Shapiro-Wilk test
    test_value, p_value = stats.shapiro(n_df[n_df.columns[i]])
    print("Resultado do teste Shapiro_Wilk test", n_df.columns[i],"é",test_value,'valor-p',p_value)
    
    ## Kolmogorov-Smirnov test
    test_value1 = stats.kstest(n_df[n_df.columns[i]],cdf='norm')
    print("Resultado do teste Kolmogorov-Smirnov test",n_df.columns[i],'é', test_value1,"\n")
```

# Anova(Análise de Variância)

***Descobrir se ha alguma  diferença nas variáveis em mais de dois grupos***

```python
# Teste Levene para testar a homogeneidade da variância
stats.levene(n_df['CO2'].astype(int)[n_df['Temperature'].astype(int) == 0],
            n_df['CO2'].astype(int)[n_df['Temperature'].astype(int)==1],
            n_df['CO2'].astype(int)[n_df["Temperature"].astype(int)==2])
```

***Vamos ver se a diferença entre os dois grupos***

```python
#separar os dois grupos de nao ocupado e sim ocupado
ocupado = n_df[n_df["Occupancy"]==1]
noocupado = n_df[n_df["Occupancy"]==0]
```

```python
# Para testar a normalidade, usamos QQ-plot, pois o teste de shapiro é afetado pelos dados levemente distorcidos
stats.probplot(ocupado['Light'],plot=plt)

#titulo
plt.title("Q-Q plot da diferença entre ocupado e luminosidade")
```

```python
# Teste Levene para homogeneidade da variância
print(stats.levene(ocupado['Light'], noocupado['Light']))
```

***interpretacão***

O valor-p é inferior a 0,05, o que significa qua não ha homogeneidade da variância entre não ocupado e ocupado

O teste de normalidade e homogeneidade da suposição de variância é violado de acordo com o QQplot e o teste de Levene, respectivamente.

Portanto, usaremos o Teste Mannwhitney_U (teste não paramétrico).

```python
# Teste mannwhitney_U 
d_1, d_2 = ocupado['CO2'], noocupado['CO2']

stat, p = stats.mannwhitneyu(d_1, d_2)
print("Teste mannwhitney_U",stat,'\nValor-p',p)
```

***Interpretação***

Os resultados mostram que o valor-p é inferior a 0,05, o que significa que há uma diferença significativa no CO2 entre grupos não ocupados(variavel) e ocupado(variavel).

Vamos ver o que acontece se usarmos o Teste t de Student (que não deveria ser usado pois as suposições de normalidade foram violadas nos dados).

```python
#Teste T de Student
stat, p = stats.ttest_ind(noocupado['CO2'], ocupado['CO2'])
desc, results = rp.ttest(noocupado['CO2'], ocupado['CO2'])
print("Valor-p: ",p)
results
```

***Teste de dependencia do mesmo grupo***

```python
ocupado['n_grupo'] = ocupado['CO2'] - ocupado['Light']
stats.probplot(ocupado['n_grupo'],plot=plt)
plt.title("Q-QPlot da diferença de CO2 e Light")
```

***Interpretação:***

O gráfico acima mostra que eles não são normalmente distribuídos, o que viola a suposição do Teste t de Student (paramétrico) pareado.

Portanto, devemos testar usando o teste Wilcoxon signed rank test.

```python
stats.wilcoxon(ocupado['n_grupo'])
```

**Interpretação:**

O valor-p mais baixo sugere que há diferença significativa entre CO2 e Luminosidade no grupo de ocupados(variavel). Ou seja, essas duas variáveis influenciam a descoberto no ambiente em estar ou nao ocupado.

***RadViz***  https://www.scikit-yb.org/en/latest/api/features/radviz.html#radviz-visualizer

RadViz é um algoritmo de visualização de dados multivariados que plota cada dimensão de recurso uniformemente ao redor da 

circunferência de um círculo e, em seguida, plota pontos no interior do círculo de modo que o ponto normalize seus valores nos 

eixos do centro para cada arco. Esse mecanismo permite tantas dimensões quantas cabem facilmente em um círculo, expandindo 

muito a dimensionalidade da visualização.

```python
x = n_df.iloc[:,:-1]
x.head()
```

```python
y = n_df["Occupancy"].values
y
```

```python
classes = ["Nao Ocupado", "Ocupado"]

vizualizer = RadViz(classes=classes)
vizualizer.fit(x,y)
vizualizer.transform(x)
vizualizer.show()
```
